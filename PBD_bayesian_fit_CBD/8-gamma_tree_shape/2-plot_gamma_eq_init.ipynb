```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.cbook import boxplot_stats
import scipy.stats
plt.style.use("customstyle")

from statsmodels.stats.multitest import multipletests


colors = ["#" + x for x in ["000000","cf5c36","bcd696","985f99","9684a1"]]
lcolors = ["#" + x for x in ["000000","e7af9c","cfb4d0","cec5d3"]] # light colors (for fill)
```

```python
def stars_pvalue(p):
    if p <= 0.001:
        return '***'
    if p <= 0.01:
        return '**'
    if p <= 0.05:
        return '*'
    return 'ns'

width = lambda p, w: 10**(np.log10(p)+w/2.)-10**(np.log10(p)-w/2.)
```

Plot $\gamma$ for simulations with $\lambda_1 = \lambda_3$.

```python
outdir = "C:/Users/pveron/Output_clusters/PBD_analog/12183"

simul_infer = pd.read_csv(outdir + "/all_simulations_inference.csv")

summary_all_mcmc = pd.read_csv(outdir + "/summary_all_MCMC.csv")

n_rep = 1 + max(simul_infer.replicate)
n_val = max(simul_infer.i_param_var)
tree_stats = pd.read_csv(outdir + "/strees_stats.csv")

simBD_tree_stats = pd.read_csv(outdir + "/simulated_BD_trees_stats.csv")
```

```python
fig, axes = plt.subplots()

PBD_names = ["PBD." + s for s in ["l1", "l2", "l3", "mu1", "mu2"]]


# Define boxplot styles 
c, lc = colors[1], lcolors[1] # for PBD boxplots
p = dict(color = c)
bp_stl_pbd = dict(capprops = p, boxprops = dict(edgecolor = c, facecolor = lc), 
                  flierprops = dict(markeredgecolor = c, markersize = 4), 
                  medianprops = p, whiskerprops = p, patch_artist = True)

c, lc = colors[3], lcolors[3] # for BD boxplots
p = dict(color = c)
bp_stl_bd = dict(capprops = p, boxprops = dict(edgecolor = c, facecolor = lc), 
                  flierprops = dict(markeredgecolor = c, markersize = 4), 
                  medianprops = p, whiskerprops = p, patch_artist = True)

w = 0.06 # width of the boxplots
d = np.exp(0.11) # distance between boxplots (in log scale)

# Statistical tests
all_pvals = np.zeros(n_val)
all_mwu = np.zeros(n_val) # Mann-Whitney U tests
n_trees = np.zeros((n_val, 2)) # Store the number of trees 

positions = np.zeros(n_val) # x positions of the plots
maxval = np.zeros((n_val, 2)) # maximum of gamma maxval[i_par, i_val, j] with j = 0 for PBD | 1 for BD


axes.axhline(0, color = "k", lw = 0.7, ls = "--")

boxes = [] # store the boxes data
boxes_bd = []

for i_val in range(n_val):
    # select the corresponding lines in the different dataframes 
    params = simul_infer.loc[(simul_infer.i_param_var == i_val + 1) &
                             (simul_infer.replicate == 0), PBD_names].to_dict("records")[0]
    tree_stats_config = tree_stats.loc[tree_stats.i_param_var == i_val + 1]
    bd_stats_config = simBD_tree_stats.loc[simBD_tree_stats.i_param_var == i_val + 1]
    
    positions[i_val] = params["PBD.l1"]

    gamma = tree_stats_config.gamma
    gamma = gamma.loc[(np.isfinite(gamma))] # remove nan
    maxval[i_val, 0] = np.max(gamma)

    gamma_bd = bd_stats_config.gamma
    gamma_bd = gamma_bd.loc[(np.isfinite(gamma_bd))] # remove nan
    maxval[i_val, 1] = np.max(gamma_bd)

    n_trees[i_val, :] = (len(gamma), len(gamma_bd))

    mwu = scipy.stats.mannwhitneyu(gamma, gamma_bd) # Mann-Whitney U test
    all_mwu[i_val] = mwu.statistic
    all_pvals[i_val] = mwu.pvalue

    boxes.append(boxplot_stats(gamma, labels = None)[0])
    boxes_bd.append(boxplot_stats(gamma_bd, labels = None)[0])
    
# Draw boxplots
x = positions
bxp = axes.bxp(boxes, positions = x / d, widths = width(x / d, w), **bp_stl_pbd)
bxp = axes.bxp(boxes_bd, positions = x * d, widths = width(x * d, w), **bp_stl_bd)

# Bonferroni correction for multiple test
bonferroni = multipletests(all_pvals, method = "bonferroni")
p_vals_corrected = bonferroni[1]

# Set ylim
ylim_lock = (-110, 30)

# Legend
axes.plot([],[], lw = 2, label = "PBD", color = bp_stl_pbd["whiskerprops"]["color"])
axes.plot([],[], lw = 2, label = "BD", color = bp_stl_bd["whiskerprops"]["color"])
axes.legend()

# Add pvalues to the graph
caphi = 15 # Position of the upper cap for the signif. bars 
def symlog_d(y): # transform to have a dy constant in symlog scale
    if y > 0:
        return y * 1.5
    return y / 1.5


for i_val in range(5):
    axes.text(x = positions[i_val], y = caphi*1.05, 
                     s = stars_pvalue(p_vals_corrected[i_val]),
                     ha = "center", size = 'small')
    # signif bars
    x = [positions[i_val] / d] * 2 + [positions[i_val] * d] * 2
    y = [symlog_d(maxval[i_val, 0]),caphi,caphi, symlog_d(maxval[i_val, 1])]
    axes.plot(x, y, lw = 1, color = "k")
axes.set_xscale("log")

axes.set_xlabel("$\lambda_1 = \lambda_3$")


axes.set_ylabel("$\gamma$ statistics")
axes.set_yscale("symlog")
axes.set_ylim(ylim_lock)

plt.savefig("../../fig/CBD_gamma_vs_BD_gamma_eq_init.pdf")

# Save the results of the MannWhitney_U_test
df = pd.DataFrame.from_dict(dict(i_param_var = np.arange(1, 1+n_val),
                                 n_trees_pbd = n_trees[:,0],
                                 n_trees_bd = n_trees[:,1], 
                                 statistics = all_mwu, 
                                 p_value = all_pvals, 
                                 corrected_p_value = p_vals_corrected))
df.to_csv("Gamma_PBD_BD_MannWhitney_U_test_eq_init.csv");
```
